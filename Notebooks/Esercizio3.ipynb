```python
import pandas as pd 
import seaborn as sns
import matplotlib as plt

primo = pd.read_csv("../Datasets/Datasets3/C_17_dataset_18_0_upFile.csv", sep=';', encoding="latin-1")
secondo = pd.read_csv("../Datasets/Datasets3/C_17_dataset_17_0_upFile.csv",sep=';', encoding="latin-1")
```

```python
for i in primo["Totale posti letto"]:
    if i== "N.D.":
        primo["Totale posti letto"]= primo["Totale posti letto"].replace(i,0)
```

```python
primo["Totale posti letto"].astype(int)
```

```python
primo["Totale posti letto"] = primo["Totale posti letto"].astype(int)
```

```python
primo["Totale posti letto"].dtypes
```

```python
secondo["Totale posti letto"]=secondo["Totale posti letto"].astype(int)
```

```python
secondo["Totale posti letto"].dtypes
```

```python
lista = [primo,secondo]

Combinazione = pd.concat(lista)
```

```python
for i in Combinazione["Descrizione Regione"]:
    if i == "LAZIO                ":
        Combinazione["Descrizione Regione"] = Combinazione["Descrizione Regione"].replace(i,"LAZIO")
```

```python
letti = Combinazione["Totale posti letto"].hist(by=Combinazione["Anno"], bins=50,figsize=(20, 15))
```

```python
sns.relplot(data=Combinazione, x="Anno", y="Totale posti letto", row="Descrizione Regione", kind="line")
```
